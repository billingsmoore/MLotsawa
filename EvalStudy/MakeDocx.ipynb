```python
import pandas as pd

df = pd.read_excel('eval_study_data_part4.ods', engine='odf')
df
```

```python
from docx import Document

doc = Document()
doc.add_heading("Evaluation Samples", level=1)
doc.add_paragraph("Each entry contains the source text, human reference, machine translation, and a space for error notes. You may add as many errors as you need. Please see the Evaluator Quickstart guide for detailed instructions.\n\n")
doc.add_paragraph("Evaluator Name: \n\n")

for _, row in df.iterrows():
    # Create a new paragraph for each entry
    p = doc.add_paragraph()

    # Add each label/value pair as separate runs so labels can be bold
    p.add_run("Source: ").bold = True
    p.add_run(f"{row['bo']}\n\n")

    p.add_run("Human Reference: ").bold = True
    p.add_run(f"{row['en']}\n\n")

    p.add_run("Machine Translation: ").bold = True
    p.add_run(f"{row['translation']}\n\n")

    p.add_run("Error: \n\n").bold = True
    # Leave the error value blank or fill in if you have one

    # Optional blank line for spacing between entries
    doc.add_paragraph()

doc.save("evaluation_set_4.docx")
print("Saved to translation_entries.docx")
```
