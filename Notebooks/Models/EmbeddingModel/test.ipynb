```python
from datasets import load_dataset

ds = load_dataset('billingsmoore/LotsawaHouse-bo-en', split='train').select(range(1000))
ds
```

```python
from easy_text_clustering.clusterer import ClusterClassifier

cc = ClusterClassifier(embed_device='cuda', summary_create=False)
cc.fit(ds['en'])
cc.show()
```

```python
from easy_text_clustering.clusterer import ClusterClassifier

cc = ClusterClassifier(embed_model_name='billingsmoore/minilm-bo', embed_device='cuda', summary_create=False)
cc.fit(ds['en'])
cc.show()
```

```python
from easy_text_clustering.clusterer import ClusterClassifier

cc = ClusterClassifier(embed_model_name='billingsmoore/minilm-bo', embed_device='cuda', summary_create=False)
cc.fit(ds['bo'])
cc.show()
```
