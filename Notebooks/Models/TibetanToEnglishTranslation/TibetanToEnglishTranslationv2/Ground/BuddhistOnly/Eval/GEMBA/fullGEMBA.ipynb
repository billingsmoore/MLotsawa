```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math

def margin_of_error(std_dev, n, confidence_level=0.95):
    """Calculate margin of error for a given sample size"""
    z_score = stats.norm.ppf((1 + confidence_level) / 2)
    return z_score * std_dev / math.sqrt(n)

def required_sample_size(std_dev, desired_margin, confidence_level=0.95):
    """Calculate required sample size for desired precision"""
    z_score = stats.norm.ppf((1 + confidence_level) / 2)
    return math.ceil((z_score * std_dev / desired_margin) ** 2)

def simulate_sampling_distribution(true_mean, true_std, sample_size, n_simulations=10000):
    """Simulate what happens when you repeatedly sample from the population"""
    sample_means = []
    for _ in range(n_simulations):
        sample = np.random.normal(true_mean, true_std, sample_size)
        sample_means.append(np.mean(sample))
    return np.array(sample_means)

# Your current small model data
small_scores = small[:500]

# Current estimates (these are our best estimates of the true parameters)
current_mean = np.mean(small_scores)
current_std = np.std(small_scores, ddof=1)  # Sample standard deviation
current_n = len(small_scores)

print("=== CURRENT SAMPLE ESTIMATES ===")
print(f"Sample size: {current_n}")
print(f"Sample mean: {current_mean:.2f}")
print(f"Sample std dev: {current_std:.2f}")

# Current precision
current_margin = margin_of_error(current_std, current_n)
print(f"Current 95% margin of error: ±{current_margin:.2f} points")
print(f"95% Confidence Interval: [{current_mean - current_margin:.2f}, {current_mean + current_margin:.2f}]")

print("\n=== PRECISION ANALYSIS FOR n=500 ===")

# What precision would n=500 give us?
n_500_margin = margin_of_error(current_std, 500)
print(f"With n=500, margin of error: ±{n_500_margin:.2f} points")
print(f"95% CI would be: [{current_mean - n_500_margin:.2f}, {current_mean + n_500_margin:.2f}]")

# Compare different sample sizes
sample_sizes = [30, 50, 100, 200, 500, 1000, 2000]
margins = [margin_of_error(current_std, n) for n in sample_sizes]

print(f"\nPrecision at different sample sizes:")
for n, margin in zip(sample_sizes, margins):
    print(f"n={n:4d}: ±{margin:.2f} points (CI width: {2*margin:.2f})")

print("\n=== SAMPLE SIZE REQUIREMENTS ===")

# What sample size is needed for different levels of precision?
desired_precisions = [0.5, 1.0, 1.5, 2.0, 3.0, 5.0]
required_ns = [required_sample_size(current_std, precision) for precision in desired_precisions]

print("Sample size needed for different precision levels:")
for precision, n in zip(desired_precisions, required_ns):
    print(f"±{precision:.1f} points: n={n:4d}")

print(f"\n500 samples gives you ±{n_500_margin:.2f} point precision")
if n_500_margin <= 2.0:
    print("✓ This is quite good precision for most practical purposes")
else:
    print("⚠ Consider if this precision meets your needs")

# Visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. Sample size vs margin of error
ax1.plot(sample_sizes, margins, 'bo-', linewidth=2, markersize=8)
ax1.axhline(y=2.0, color='red', linestyle='--', alpha=0.7, label='±2 point target')
ax1.axhline(y=1.0, color='green', linestyle='--', alpha=0.7, label='±1 point target')
ax1.axvline(x=500, color='orange', linestyle='--', alpha=0.7, label='n=500')
ax1.set_xlabel('Sample Size')
ax1.set_ylabel('Margin of Error (±points)')
ax1.set_title('Precision vs Sample Size')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xscale('log')

# 2. Confidence interval width vs sample size
ci_widths = [2 * margin for margin in margins]
ax2.plot(sample_sizes, ci_widths, 'ro-', linewidth=2, markersize=8)
ax2.axhline(y=4.0, color='red', linestyle='--', alpha=0.7, label='4 point CI width')
ax2.axhline(y=2.0, color='green', linestyle='--', alpha=0.7, label='2 point CI width')
ax2.axvline(x=500, color='orange', linestyle='--', alpha=0.7, label='n=500')
ax2.set_xlabel('Sample Size')
ax2.set_ylabel('95% CI Width (points)')
ax2.set_title('Confidence Interval Width vs Sample Size')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_xscale('log')

# 3. Simulate sampling distribution for n=500
simulated_means_500 = simulate_sampling_distribution(current_mean, current_std, 500, 1000)
ax3.hist(simulated_means_500, bins=30, alpha=0.7, density=True, color='skyblue')
ax3.axvline(current_mean, color='red', linestyle='-', linewidth=2, label=f'True mean ≈ {current_mean:.1f}')
ax3.axvline(current_mean - n_500_margin, color='orange', linestyle='--', label=f'95% CI bounds')
ax3.axvline(current_mean + n_500_margin, color='orange', linestyle='--')
ax3.set_xlabel('Sample Mean')
ax3.set_ylabel('Density')
ax3.set_title('Distribution of Sample Means (n=500)')
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. Show how estimate improves with sample size
np.random.seed(42)  # For reproducibility
true_population = np.random.normal(current_mean, current_std, 10000)
cumulative_means = []
cumulative_margins = []

for i in range(10, min(2000, len(true_population))):
    sample = true_population[:i]
    cum_mean = np.mean(sample)
    cum_std = np.std(sample, ddof=1)
    cum_margin = margin_of_error(cum_std, i)
    cumulative_means.append(cum_mean)
    cumulative_margins.append(cum_margin)

sample_range = range(10, len(cumulative_means) + 10)
ax4.plot(sample_range, cumulative_means, 'b-', linewidth=2, label='Cumulative Mean')
ax4.fill_between(sample_range, 
                np.array(cumulative_means) - np.array(cumulative_margins),
                np.array(cumulative_means) + np.array(cumulative_margins),
                alpha=0.3, color='blue', label='95% CI')
ax4.axhline(current_mean, color='red', linestyle='--', alpha=0.7, label='True Mean')
ax4.axvline(500, color='orange', linestyle='--', alpha=0.7, label='n=500')
ax4.set_xlabel('Sample Size')
ax4.set_ylabel('Estimated Mean')
ax4.set_title('Mean Estimate Convergence')
ax4.legend()
ax4.grid(True, alpha=0.3)
ax4.set_ylim(current_mean - 10, current_mean + 10)

plt.tight_layout()
plt.show()

print("\n=== ASSESSMENT FOR n=500 ===")

# Calculate what percentage of the standard deviation our margin represents
precision_as_pct_of_std = (n_500_margin / current_std) * 100

print(f"With n=500:")
print(f"• Margin of error: ±{n_500_margin:.2f} points")
print(f"• This is {precision_as_pct_of_std:.1f}% of the standard deviation")
print(f"• 95% confidence interval width: {2*n_500_margin:.2f} points")

if n_500_margin <= 1.5:
    assessment = "EXCELLENT"
elif n_500_margin <= 2.5:
    assessment = "GOOD"
elif n_500_margin <= 4.0:
    assessment = "ADEQUATE"
else:
    assessment = "CONSIDER MORE SAMPLES"

print(f"• Assessment: {assessment} precision")

print(f"\nContext:")
print(f"• Your scores range from {min(small_scores)} to {max(small_scores)} (range: {max(small_scores) - min(small_scores)})")
print(f"• A ±{n_500_margin:.1f} point margin is {(n_500_margin/(max(small_scores) - min(small_scores)))*100:.1f}% of your score range")

print(f"\n=== RECOMMENDATION ===")
if n_500_margin <= 2.0:
    print("✓ 500 samples should give you excellent precision for estimating the true mean")
    print("✓ This level of precision is sufficient for most research and practical applications")
else:
    print("⚠ Consider whether ±{:.1f} point precision meets your requirements".format(n_500_margin))
    print(f"⚠ For better precision, consider n={required_sample_size(current_std, 2.0)}")

print(f"\nNote: These calculations assume your current sample is representative of the population.")
print(f"The true population standard deviation might differ from your sample estimate.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math

def margin_of_error(std_dev, n, confidence_level=0.95):
    """Calculate margin of error for a given sample size"""
    z_score = stats.norm.ppf((1 + confidence_level) / 2)
    return z_score * std_dev / math.sqrt(n)

def required_sample_size(std_dev, desired_margin, confidence_level=0.95):
    """Calculate required sample size for desired precision"""
    z_score = stats.norm.ppf((1 + confidence_level) / 2)
    return math.ceil((z_score * std_dev / desired_margin) ** 2)

def simulate_sampling_distribution(true_mean, true_std, sample_size, n_simulations=10000):
    """Simulate what happens when you repeatedly sample from the population"""
    sample_means = []
    for _ in range(n_simulations):
        sample = np.random.normal(true_mean, true_std, sample_size)
        sample_means.append(np.mean(sample))
    return np.array(sample_means)

# Your current small model data
small_scores = base[:500]

# Current estimates (these are our best estimates of the true parameters)
current_mean = np.mean(small_scores)
current_std = np.std(small_scores, ddof=1)  # Sample standard deviation
current_n = len(small_scores)

print("=== CURRENT SAMPLE ESTIMATES ===")
print(f"Sample size: {current_n}")
print(f"Sample mean: {current_mean:.2f}")
print(f"Sample std dev: {current_std:.2f}")

# Current precision
current_margin = margin_of_error(current_std, current_n)
print(f"Current 95% margin of error: ±{current_margin:.2f} points")
print(f"95% Confidence Interval: [{current_mean - current_margin:.2f}, {current_mean + current_margin:.2f}]")

print("\n=== PRECISION ANALYSIS FOR n=500 ===")

# What precision would n=500 give us?
n_500_margin = margin_of_error(current_std, 500)
print(f"With n=500, margin of error: ±{n_500_margin:.2f} points")
print(f"95% CI would be: [{current_mean - n_500_margin:.2f}, {current_mean + n_500_margin:.2f}]")

# Compare different sample sizes
sample_sizes = [30, 50, 100, 200, 500, 1000, 2000]
margins = [margin_of_error(current_std, n) for n in sample_sizes]

print(f"\nPrecision at different sample sizes:")
for n, margin in zip(sample_sizes, margins):
    print(f"n={n:4d}: ±{margin:.2f} points (CI width: {2*margin:.2f})")

print("\n=== SAMPLE SIZE REQUIREMENTS ===")

# What sample size is needed for different levels of precision?
desired_precisions = [0.5, 1.0, 1.5, 2.0, 3.0, 5.0]
required_ns = [required_sample_size(current_std, precision) for precision in desired_precisions]

print("Sample size needed for different precision levels:")
for precision, n in zip(desired_precisions, required_ns):
    print(f"±{precision:.1f} points: n={n:4d}")

print(f"\n500 samples gives you ±{n_500_margin:.2f} point precision")
if n_500_margin <= 2.0:
    print("✓ This is quite good precision for most practical purposes")
else:
    print("⚠ Consider if this precision meets your needs")

# Visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. Sample size vs margin of error
ax1.plot(sample_sizes, margins, 'bo-', linewidth=2, markersize=8)
ax1.axhline(y=2.0, color='red', linestyle='--', alpha=0.7, label='±2 point target')
ax1.axhline(y=1.0, color='green', linestyle='--', alpha=0.7, label='±1 point target')
ax1.axvline(x=500, color='orange', linestyle='--', alpha=0.7, label='n=500')
ax1.set_xlabel('Sample Size')
ax1.set_ylabel('Margin of Error (±points)')
ax1.set_title('Precision vs Sample Size')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xscale('log')

# 2. Confidence interval width vs sample size
ci_widths = [2 * margin for margin in margins]
ax2.plot(sample_sizes, ci_widths, 'ro-', linewidth=2, markersize=8)
ax2.axhline(y=4.0, color='red', linestyle='--', alpha=0.7, label='4 point CI width')
ax2.axhline(y=2.0, color='green', linestyle='--', alpha=0.7, label='2 point CI width')
ax2.axvline(x=500, color='orange', linestyle='--', alpha=0.7, label='n=500')
ax2.set_xlabel('Sample Size')
ax2.set_ylabel('95% CI Width (points)')
ax2.set_title('Confidence Interval Width vs Sample Size')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_xscale('log')

# 3. Simulate sampling distribution for n=500
simulated_means_500 = simulate_sampling_distribution(current_mean, current_std, 500, 1000)
ax3.hist(simulated_means_500, bins=30, alpha=0.7, density=True, color='skyblue')
ax3.axvline(current_mean, color='red', linestyle='-', linewidth=2, label=f'True mean ≈ {current_mean:.1f}')
ax3.axvline(current_mean - n_500_margin, color='orange', linestyle='--', label=f'95% CI bounds')
ax3.axvline(current_mean + n_500_margin, color='orange', linestyle='--')
ax3.set_xlabel('Sample Mean')
ax3.set_ylabel('Density')
ax3.set_title('Distribution of Sample Means (n=500)')
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. Show how estimate improves with sample size
np.random.seed(42)  # For reproducibility
true_population = np.random.normal(current_mean, current_std, 10000)
cumulative_means = []
cumulative_margins = []

for i in range(10, min(2000, len(true_population))):
    sample = true_population[:i]
    cum_mean = np.mean(sample)
    cum_std = np.std(sample, ddof=1)
    cum_margin = margin_of_error(cum_std, i)
    cumulative_means.append(cum_mean)
    cumulative_margins.append(cum_margin)

sample_range = range(10, len(cumulative_means) + 10)
ax4.plot(sample_range, cumulative_means, 'b-', linewidth=2, label='Cumulative Mean')
ax4.fill_between(sample_range, 
                np.array(cumulative_means) - np.array(cumulative_margins),
                np.array(cumulative_means) + np.array(cumulative_margins),
                alpha=0.3, color='blue', label='95% CI')
ax4.axhline(current_mean, color='red', linestyle='--', alpha=0.7, label='True Mean')
ax4.axvline(500, color='orange', linestyle='--', alpha=0.7, label='n=500')
ax4.set_xlabel('Sample Size')
ax4.set_ylabel('Estimated Mean')
ax4.set_title('Mean Estimate Convergence')
ax4.legend()
ax4.grid(True, alpha=0.3)
ax4.set_ylim(current_mean - 10, current_mean + 10)

plt.tight_layout()
plt.show()

print("\n=== ASSESSMENT FOR n=500 ===")

# Calculate what percentage of the standard deviation our margin represents
precision_as_pct_of_std = (n_500_margin / current_std) * 100

print(f"With n=500:")
print(f"• Margin of error: ±{n_500_margin:.2f} points")
print(f"• This is {precision_as_pct_of_std:.1f}% of the standard deviation")
print(f"• 95% confidence interval width: {2*n_500_margin:.2f} points")

if n_500_margin <= 1.5:
    assessment = "EXCELLENT"
elif n_500_margin <= 2.5:
    assessment = "GOOD"
elif n_500_margin <= 4.0:
    assessment = "ADEQUATE"
else:
    assessment = "CONSIDER MORE SAMPLES"

print(f"• Assessment: {assessment} precision")

print(f"\nContext:")
print(f"• Your scores range from {min(small_scores)} to {max(small_scores)} (range: {max(small_scores) - min(small_scores)})")
print(f"• A ±{n_500_margin:.1f} point margin is {(n_500_margin/(max(small_scores) - min(small_scores)))*100:.1f}% of your score range")

print(f"\n=== RECOMMENDATION ===")
if n_500_margin <= 2.0:
    print("✓ 500 samples should give you excellent precision for estimating the true mean")
    print("✓ This level of precision is sufficient for most research and practical applications")
else:
    print("⚠ Consider whether ±{:.1f} point precision meets your requirements".format(n_500_margin))
    print(f"⚠ For better precision, consider n={required_sample_size(current_std, 2.0)}")

print(f"\nNote: These calculations assume your current sample is representative of the population.")
print(f"The true population standard deviation might differ from your sample estimate.")
```
