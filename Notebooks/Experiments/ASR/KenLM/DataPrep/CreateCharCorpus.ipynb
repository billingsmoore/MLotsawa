```python
def tibetan_char_tokenize(s: str) -> str:
    # remove punctuation not needed for LM
    s = s.replace("།", " ").replace("༎", " ").replace("་", " ")
    s = "".join(ch if "\u0f00" <= ch <= "\u0fff" else " " for ch in s)
    chars = [c for c in s if c.strip()]
    return " ".join(chars)
```

```python
import re
from tqdm import tqdm

input_file = "../Data/all_sentences.txt"
output_file = "../Data/char_lm_corpus.txt"

def clean_tibetan_text(text):
    # normalize spacing
    text = re.sub(r"\s+", " ", text).strip()
    return text

with open(input_file, "r") as fin, open(output_file, "w") as fout:
    for line in tqdm(fin):
        cleaned = clean_tibetan_text(line)
        tokenized = tibetan_char_tokenize(cleaned)
        if tokenized.strip():
            fout.write(tokenized + "\n")
```
