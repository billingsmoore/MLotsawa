```python
import re
from tqdm import tqdm
from tibetan_wer.metrics import word_segment as segment

def tibetan_word_tokenize(text: str) -> str:
    # Replace Tibetan punctuation with spaces
    text = text.replace("།", " ").replace("༎", " ")

    # Keep only Tibetan range; replace everything else with space
    text = "".join(ch if "\u0f00" <= ch <= "\u0fff" else " " for ch in text)

    # Clean spacing before segmentation
    text = re.sub(r"\s+", " ", text).strip()

    # Segment into words (provided by your Tibetan segmenter)
    words = segment(text)            # <--- main change

    # Filter out any empty/whitespace tokens
    words = [w.strip() for w in words if w.strip()]

    return " ".join(words)


def clean_tibetan_text(text: str) -> str:
    return re.sub(r"\s+", " ", text).strip()

```

```python
input_file = "../Data/all_sentences.txt"
output_file = "../Data/word_lm_corpus.txt"

with open(input_file, "r", encoding="utf-8") as fin, \
     open(output_file, "w", encoding="utf-8") as fout:

    for line in tqdm(fin):
        cleaned = clean_tibetan_text(line)
        tokenized = tibetan_word_tokenize(cleaned)
        if tokenized:
            fout.write(tokenized + "\n")
```
