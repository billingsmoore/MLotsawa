```python
import re

def tibetan_syllable_tokenize(s: str) -> str:
    # Replace Tibetan punctuation with spaces
    s = s.replace("།", " ").replace("༎", " ")
    
    # Keep Tibetan Unicode range; replace others with space
    s = "".join(ch if "\u0f00" <= ch <= "\u0fff" else " " for ch in s)

    # Split on tsek (་) to get syllables
    syllables = s.split("་")
    
    # Remove empty entries & strip whitespace
    syllables = [syll.strip() for syll in syllables if syll.strip()]

    return " ".join(syllables)


def clean_tibetan_text(text):
    return re.sub(r"\s+", " ", text).strip()

```

```python
input_file = "../Data/all_sentences.txt"
output_file = "../Data/syl_lm_corpus.txt"

from tqdm import tqdm


with open(input_file, "r", encoding="utf-8") as fin, open(output_file, "w", encoding="utf-8") as fout:
    for line in tqdm(fin):
        cleaned = clean_tibetan_text(line)
        tokenized = tibetan_syllable_tokenize(cleaned)
        if tokenized:
            fout.write(tokenized + "\n")
```
